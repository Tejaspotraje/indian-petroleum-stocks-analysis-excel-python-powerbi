```python
import pandas as pd
import numpy as np
from tabulate import tabulate

Reliance_Industries_path = "D:\\Indian Peterolium Stocks Analysis Project\\Input File\\Piviot file\\Reliance Industry Limted (Piviot).xlsx"
df = pd.read_excel(Reliance_Industries_path)
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
df.rename(columns={"date_1": "date"}, inplace=True)
```

```python
df = df.rename(columns={
    'no.of_share': 'no_of_share',
    'no._of_trades': 'no_of_trades',
    'total_turnover_(rs.)': 'total_turnover_(Rs)',
    '%_deli._qty_to_traded_qty': '%_deli_qty_to_traded_qty',
    'spread_high-low': 'spread_(high-low)',
    'spread_close-open': 'spread_(close-open)',
    'no.of_shares': 'no_of_shares',
    'PE_Ratio': 'pe_ratio'
})
```

```python
df['year'] = df['year'].astype(int)
df['open_price'] = df['open_price'].astype(float)
df['high_price'] = df['high_price'].astype(float)
df['low_price'] = df['low_price'].astype(float)
df['close_price'] = df['close_price'].astype(float)
df['%_deli_qty_to_traded_qty'] = df['%_deli_qty_to_traded_qty'].astype(float)
df['spread_(high-low)'] = df['spread_(high-low)'].astype(float)
df['spread_(close-open)'] = df['spread_(close-open)'].astype(float)
df['eps'] = df['eps'].astype(float)
df['price'] = df['price'].astype(float)
df['opm'] = df['opm'].astype(float)
```

```python
df = df.drop_duplicates()
```

```python
df['pe_ratio'] = df['price'] / df['eps']

face_value = 10

df['outstanding_shares'] = (df['equity_share_capital'] * 1e7) / face_value
df['market_cap'] = df['price'] * df['outstanding_shares']
```

```python
df.to_excel("D:/Indian Peterolium Stocks Analysis Project/Output File/Reliance Clean ETL Output File.xlsx", index=False)
df.to_csv("D:/Indian Peterolium Stocks Analysis Project/Output File/Reliance Clean ETL Output File.csv", index=False)
```
