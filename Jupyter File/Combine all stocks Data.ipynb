```python
import pandas as pd
```

```python
# Load files
df_reliance = pd.read_excel("D:\Indian Peterolium Stocks Analysis Project\Output File\Excel file\Reliance Clean ETL Output File.xlsx")
df_bharat = pd.read_excel("D:\Indian Peterolium Stocks Analysis Project\Output File\Excel file\Bharat Oil Clean ETL Output File.xlsx")
df_hindustan = pd.read_excel("D:\Indian Peterolium Stocks Analysis Project\Output File\Excel file\Hindustan Oil Clean ETL Output File.xlsx")
df_ongc = pd.read_excel("D:\Indian Peterolium Stocks Analysis Project\Output File\Excel file\ONGC Clean ETL Output File.xlsx")
df_oil = pd.read_excel("D:\Indian Peterolium Stocks Analysis Project\Output File\Excel file\Oil Clean ETL Output File.xlsx")   
df_indian = pd.read_excel("D:\Indian Peterolium Stocks Analysis Project\Output File\Excel file\Indian Oil Clean ETL Output File.xlsx")

# Append (stack rows)
combined_df = pd.concat(
    [df_reliance, df_bharat, df_hindustan, df_ongc, df_oil, df_indian],
    ignore_index=True
)


# Add sector column
combined_df['sector'] = combined_df['companies'].apply(
    lambda x: 'Private' if 'Reliance' in x else 'Public'
)

# Save final dataset
combined_df.to_csv("D:\\Indian Peterolium Stocks Analysis Project\\Output File\\CSV file\\Indian Petroleum Stocks Combined.csv", index=False)
combined_df.to_excel("D:\\Indian Peterolium Stocks Analysis Project\\Output File\\Excel File\\Indian Petroleum Stocks Combined.xlsx", index=False)

combined_df


```
