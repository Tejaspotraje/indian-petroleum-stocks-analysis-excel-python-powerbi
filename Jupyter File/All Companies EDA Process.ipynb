### 1️⃣ Which company has the highest average closing price?

```python
df.groupby('companies')['close_price'].mean().sort_values(ascending=False)
```

```python
df.groupby('companies')['spread_(high-low)'].mean()
```

```python
df.pivot(index='year', columns='companies', values='close_price')
```

```python
df.groupby('year')['total_turnover_(Rs)'].sum()
```
